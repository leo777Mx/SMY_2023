## 1. Discretización de ecuaciones diferenciales

Partiendo de la siguiente ecuación diferencial con parte temporal. (1), se busca discretizar mediente los esquemas implicito, semi-implicito y explicito.    
<br>
<br>

$$\frac{\partial^2P}{\partial{r^2}} + \frac{1}{r}\frac{\partial{P}}{\partial{r}} = \frac{\mu\phi C_{t}}{k} \frac{\partial{P}}{\partial{t}} \quad ...(1)$$ 

<img src="./utils/figs/fig1.PNG"/>

<br>
Donde las unidades para viscosidad ($\mu$) y permeabilidad (k) son:

$cP*(1/psi)/mD*(psi/dia) = psi/ft^2+psi/ft^2$ \
$cP -> psi*dia$ \
$mD -> ft^2$

## 2. Discretización mediante diferencias finitas y metodo implicito 
Reescribiendo la ec. (1), queda:
<br>
$$\frac{\partial^2P}{\partial{r^2}} + \frac{1}{r}\frac{\partial{P}}{\partial{r}}- C1 \frac{\partial{P}}{\partial{t}} = 0 \quad ...(2)$$

Donde:
<br>
$C_{1}=\frac{\mu\phi C_{t}}{k}$
<br>
<br>

Se parte del esquema teta de discretizacion en el tiempo, el cual se expresa como:
<br>
$$\frac{\partial F}{\partial t} = (1-\theta)f^n(t,r)+\theta f^{n+1}(t,r) \quad ...(3)$$

Donde los superindices $n$ y $n+1$ representan el tiempo inicial y el tiempo inicial más un paso de tiempo, respectivamente.
<br>
<br>
Para discretizar implicitamente (3), $\theta$ se sustituye por el valor 1, quedando la siguiente ecuación:
<br>
$$f^{n+1}(t,r) \qquad ...(4)$$
<br>

Aplicando diferencias finitas y la ecuación (4) a (2), queda lo siguiente:
<br>

$$\frac{P^{n+1}_{i+1}-2P^{n+1}_{i}+P^{n+1}_{i-1}}{\Delta{r^2}} + \frac{1}{r}\frac{P^{n+1}_{i+1}-P^{n+1}_{i-1}}{2\Delta{r}}-C_{1}(\frac{P^{n+1}_{i}-P^n_{i}}{\Delta{t}}) = 0 \quad ...(5)$$

<br>
Agrupando terminos,

$$-P^{n+1}_{i}(\frac{2}{\Delta{r^2}}+\frac{C_{1}}{\Delta{t}}) + P^{n+1}_{i+1}(\frac{1}{\Delta{r^2}}+\frac{1}{r 2\Delta{r}}) + P^{n+1}_{i-1}(\frac{1}{\Delta{r^2}}-\frac{1}{r2\Delta{r}}) - C_{1}\frac{P^n_{i}}{\Delta{t}} = 0 \quad ...(6)$$

<br>
Simplificando la ecuación, se obtiene:
$$AP P_{i} + AE P_{i+1} + AW P_{i-1} - B = 0 \quad ...(7)$$

$Donde:$

$AP = -(\frac{2}{\Delta{r^2}} + \frac{C_{1}}{\Delta{t}})$\
$AE = \frac{1}{\Delta{r}^2}+\frac{1}{r_{i+1}2\Delta{r}}$\
$AW = \frac{1}{\Delta{r}^2}-\frac{1}{r_{i+1}2\Delta{r}}$\
$B = -C_{1} \frac{P_{old}}{\Delta{t}}$

### 2.1 Desarrollo del código 

```python
# importar bibliotecas necesarias
import matplotlib.pyplot as plt   # importa la biblioteca matplotlib con el sobrenombre plt (gráficar) 
from utils.funciones import solver as SL  # importa el archivo propio solverLin con el sobrenombre SL (resolver sistemas tridiagonales) 
import numpy as np                # importa la biblioteca numpy para utilizar matrices y vectores (numerical python)
from time import time
```

```python
# Constantes del problema
k=100.0    #Permeabilidad en mD
phi=0.1   #Porosidad 
Cr=5e-07 #1/psi
Cf=2e-06  #1/psi
Mu=50.0  #cP
q=-500.0 #rb/day
A=1000  # ft2

#Propiedades del fluido y de la roca
Ct=Cf+Cr
k=k*1.0623e-14   #mD -------> ft**2
Mu=Mu*1.45e-07   #cP------> psi*s
q=q*5.61        #rb/day ---ft**3/day
q=q/86400.0     #ft**3/day ---> ft**3/s 
C1=(Mu*phi*Ct)/k
#C2=(q*Mu*deltar)/(A*k)
```

```python
#SOLUCIÓN METODO IMPLICITO

Nr=100    #Numero de nodos del espacio discreto
Pa=2000;  Pb=3000
rw=0.5; re=500.5
tiempoSimulacion=2000 #Días
deltat=1.0  #Días
Pinicial=Pa   #Presión inicial en psi
r=np.linspace(rw,re,Nr)  #genera la variable independiente
NrInt=Nr-2   #Nodos internos donde las incognitas "p" son desconocidas
Lr=re-rw
deltar=Lr/(Nr-1)
deltat=deltat*86400.0  #dia ----> s

# coeficientes
AP=np.zeros(NrInt)
AE=np.zeros(NrInt)
AW=np.zeros(NrInt)
B=np.zeros(NrInt)
Pi=np.zeros(NrInt)
Press=np.zeros(Nr)
P_old=np.ones(NrInt)*Pinicial

valorMaxP=10.0
tolP=1.0
tiempo=0.0

while ((tiempo<tiempoSimulacion)and(valorMaxP>tolP)):    
    for i in range (0,NrInt):
        AP[i]=-(2/deltar**2+C1/deltat)
        AE[i]= 1/deltar**2+1/(r[i+1]*2*deltar)
        AW[i]= 1/deltar**2-1/(r[i+1]*2*deltar)
        B[i]=-C1*P_old[i]/deltat
        
    #Condición de frontera en rw (Pa conocida)
    B[0]=B[0]-AW[0]*Pa

    #Condición de frontera en rw (Q constate)
    # B[0]=B[0]-AW[0]*C2
    # AP[0]=AP[0]+AW[0]

    
    #Condición de frontera en re (Pb conocida)    
    B[NrInt-1]=B[NrInt-1]-AE[NrInt-1]*Pb
        
    #Solución mediante el algortimo de thomas
    Pi=SL.thomas1D(AP, -AE, -AW, B)   #
    
    # Pa=C2+Pi[0]  #Presión calculada cuando usamos una condición de frontera a q=cte

    valorMaxP=np.max(abs(P_old-Pi))  #Paro estado estable
    
    #Actualización de la presión anterior para el siguiente paso de tiempo
    P_old=np.copy(Pi)
    
    tiempo=tiempo+deltat/86400.0
    
    Press[0]=Pa
    Press[Nr-1]=Pb
    Press[1:Nr-1]=Pi

    
    plt.figure("solucion MI")
    plt.plot(r, Press)
    print("tiempo de simulación:", tiempo, "\t", "diferencia de P:",  valorMaxP)
    
plt.grid()
```

## 2. Metodo semi implicito Crank-Nicholson

Para obtener las ecuaciones del metodo semi-implicito Crank-Nicholson, se parte del esquema de discretización en el tiempo ec.(3), donde $\theta$ adquiere el valor de 0.5. Dando resultado a la siguiente ecuacion:
<br>
<br>
$$\frac{\partial F}{\partial t} = (1-0.5)f^n(t,r)+0.5 f^{n+1}(t,r) \qquad ...(8)$$

<br>
Donde la derivada de la funcion en el tiempo equivale a $1/2$ de la cuncion evaluada en el tiempo n más la funcion evaluada en el tiempo $n + 1$.
<br>
La ecuación se puede expresar de la siguiente manera:
<br>
$$2C_{1}\frac{P^{n+1}_{i} - P^n_{i}}{\Delta{t}} = $$
<br>
$$\frac{P^n_{i-1} - 2P^n_i + P^n_{i+1}}{\Delta{r^2}} + \frac{1}{r}\frac{P^n_{i+1}-P^n_{i-1}}{2\Delta{r}} + \frac{P^{n+1}_{1-1} - 2P^{n+1}_i + P^{n+1}_{i+1}}{\Delta{r^2}} + \frac{1}{r}\frac{P^{n+1}_{i+1} - P^{n+1}_{i-1}}{2\Delta{r}}  \qquad ...(9)$$

<br>
Ordenando los terminos de la ecuación (9), queda:

$$\left[-\frac{1}{\Delta{r^2}} + \frac{1}{r2\Delta{r}} \right]P^n_{i-1} + \left[\frac{2}{\Delta{r^2}} - \frac{2C_1}{\Delta{t}}  \right]P^n_{i} + \left[-\frac{1}{\Delta{r^2}} - \frac{1}{r2\Delta{r}} \right]P^n_{i+1} = $$

<br>
$$\left[\frac{1}{\Delta{r^2}} - \frac{1}{r2\Delta{r}} \right]P^{n+1}_{i-1} + \left[-\frac{2}{\Delta{r^2}} - \frac{2C_1}{\Delta{t}}  \right]P^{n+1}_{i} + \left[\frac{1}{\Delta{r^2}} + \frac{1}{r2\Delta{r}} \right]P^{n+1}_{i+1}
\qquad ...(10)$$

<br>
La ecuación (10) se puede reescribir en termino de los coeficientes como:
<br>

$$AW P^{n+1}_{i-1} + AP P^{n+1}_{i} + AEP^{n+1}_{i+1} = B
\qquad ...(11)$$

Donde:
<br>
$$AW = \frac{1}{\Delta{r^2}} - \frac{1}{r2\Delta{r}}$$
<br>
$$AP = -\frac{2}{\Delta{r^2}} - \frac{2C_1}{\Delta{t}}$$
<br>
$$AE = \frac{1}{\Delta{r^2}} + \frac{1}{r2\Delta{r}}$$
<br>
$$B = \left[-\frac{1}{\Delta{r^2}} + \frac{1}{r2\Delta{r}} \right]P^n_{i-1} + \left[\frac{2}{\Delta{r^2}} - \frac{2C_1}{\Delta{t}}  \right]P^n_{i} + \left[-\frac{1}{\Delta{r^2}} - \frac{1}{r2\Delta{r}} \right]P^n_{i+1}$$





Para asignar los valores de presión anterior (P_old) en las fronteras, se tiene la siguiente ilustración:
<br>
<img src="./utils/figs/semi_implicito_malla.PNG" />

```python
#SOLUCIÓN SEMI-IMPLICITO CRANK-NICHOLSON
Nr=100    #Numero de nodos del espacio discreto
Pa=2000;  Pb=3000;
rw=0.5; re=500.5

tiempoSimulacion=2000 #Días
deltat=1.0  #Días
Pinicial=Pa   #Presión inicial en psi
r=np.linspace(rw,re,Nr)  #genera la variable independiente
NrInt=Nr-2   #Nodos internos donde las incognitas "p" son desconocidas
Lr=re-rw
deltar=Lr/(Nr-1)
deltat=deltat*86400.0  #dia ----> s

#coeficientes
AP=np.zeros(NrInt)
AE=np.zeros(NrInt)
AW=np.zeros(NrInt)
B=np.zeros(NrInt)
Pie=np.zeros(NrInt)
Press=np.zeros(Nr)

P_old=np.ones(Nr)*Pinicial
P_old[0]=Pa
P_old[Nr-1]=Pb

valorMaxP=10.0
tolP=1
deltat=deltat/200

#Solución por el método implicito y con derivadas centrales en el espacio
tiempo=0.0
while ((tiempo<tiempoSimulacion)and(valorMaxP>tolP)):
    
    for i in range (1,Nr-1):
        GEx=(P_old[i-1]-2*P_old[i]+P_old[i+1])/deltar**2+(1/r[i])*((P_old[i+1]-P_old[i-1])/(2*deltar))
        
        AP[i-1]=-(2/deltar**2+2*C1/deltat)
        AE[i-1]= 1/deltar**2+1/(r[i]*2*deltar)
        AW[i-1]= 1/deltar**2-1/(r[i]*2*deltar)
        B[i-1]=-2*C1*P_old[i]/deltat-GEx

        
    #Condición de frontera en rw (Pa conocida)
    B[0]=B[0]-AW[0]*Pa

    #Condición de frontera en rw (Q constate)
    # B[0]=B[0]-AW[0]*C2
    # AP[0]=AP[0]+AW[0]

    
    #Condición de frontera en re (Pb conocida)    
    B[NrInt-1]=B[NrInt-1]-AE[NrInt-1]*Pb
        
    #Solución mediante el algortimo de thomas
    Pie=SL.thomas1D(AP, -AE, -AW, B)   #
    
    # Pa=C2+Pi[0]  #Presión calculada cuando usamos una condición de frontera a q=cte

    valorMaxP=np.max(abs(P_old[1:Nr-1]-Pie))  #Paro estado estable
    
    #Actualización de la presión anterior para el siguiente paso de tiempo
    P_old[1:Nr-1]=np.copy(Pie)
    
    tiempo=tiempo+deltat/86400.0
    
    Press[0]=Pa
    Press[Nr-1]=Pb
    Press[1:Nr-1]=Pie

    
    plt.figure("solucion MIe")
    plt.plot(r, Press)
    #print("tiempo de simulación:", tiempo, "\t", "diferencia de P:",  valorMaxP)
    
plt.grid()
```

## 3. Metodo Explicito y con derivadas centrales en el espacio

```python
# definir variables
Nr=100    #Numero de nodos del espacio discreto
Pa=2000;  Pb=3000;  rw=0.5; re=500.5
tiempoSimulacion=2000 #Días
deltat=1.0  #Días

k=100.0    #Permeabilidad en mD
phi=0.1   #Porosidad 
Cr=5e-07 #1/psi
Cf=2e-06  #1/psi
Mu=50.0  #cP
Pinicial=Pa   #Presión inicial en psi
q=-500.0 #rb/day
A=1000  # ft2

r=np.linspace(rw,re,Nr)  #genera la variable independiente
NrInt=Nr-2   #Nodos internos donde las incognitas "p" son desconocidas
Lr=re-rw
deltar=Lr/(Nr-1) 

#Propiedades del fluido y de la roca
Ct=Cf+Cr
k=k*1.0623e-14   #mD -------> ft**2
Mu=Mu*1.45e-07   #cP------> psi*s
deltat=deltat*86400.0  #dia ----> s
q=q*5.61        #rb/day ---ft**3/day
q=q/86400.0     #ft**3/day ---> ft**3/s 
C1=(Mu*phi*Ct)/k
C2=(q*Mu*deltar)/(A*k)

AP=np.zeros(NrInt)
AE=np.zeros(NrInt)
AW=np.zeros(NrInt)
B=np.zeros(NrInt)
Pe=np.zeros(NrInt)
P_old=np.ones(NrInt)*Pinicial

#tiempoSimulacion=1
#Solución por el método explicíto y con derivadas centrales en el espacio
deltat=deltat/1000

tiempo=0.0
while (tiempo<tiempoSimulacion):
    
    for i in range (0,NrInt):
        AP[i]=-2/deltar**2
        AE[i]= 1/deltar**2+1/(r[i+1]*2*deltar)
        AW[i]= 1/deltar**2-1/(r[i+1]*2*deltar)
        B[i]=P_old[i]
        
        if (i>0 and i<NrInt-1):
            Pe[i]=deltat/C1*(AW[i]*P_old[i-1]+AP[i]*P_old[i]+AE[i]*P_old[i+1])+B[i]
            
        #Condición de frontera en rw (Pa conocida)
        if (i==0):
            Pe[i]=deltat/C1*(AW[i]*Pa+AP[i]*P_old[i]+AE[i]*P_old[i+1])+B[i]
 
        #Condición de frontera en re (Pb conocida)    
        if (i==NrInt-1):
            Pe[i]=deltat/C1*(AW[i]*P_old[i-1]+AP[i]*P_old[i]+AE[i]*Pb)+B[i]
            
    # #Condición de frontera en rw (Pa conocida)
    # AP[0]=-2/deltar**2    
    # AE[0]=1/deltar**2+1/(r[1]*2*deltar)
    # AW[0]=1/deltar**2-1/(r[1]*2*deltar)
    # B[0]=P_old[0]
    
    # Pe[0]=deltat/C1*(AW[0]*Pa+AP[0]*P_old[0]+AE[0]*P_old[1])+B[0]
    
    # # #Condición de frontera en re (Pb conocida)    
    # AP[NrInt-1]=-2/deltar**2    
    # AE[NrInt-1]=1/deltar**2+1/(r[NrInt]*2*deltar)
    # AW[NrInt-1]=1/deltar**2-1/(r[NrInt]*2*deltar)
    # B[NrInt-1]=P_old[NrInt-1]
        
    # Pe[NrInt-1]=deltat/C1*(AW[NrInt-1]*P_old[NrInt-2]+AP[NrInt-1]*P_old[NrInt-1]+AE[NrInt-1]*Pb)+B[NrInt-1]

    # #Actualización de la presión anterior para el siguiente paso de tiempo
    P_old=np.copy(Pe)
    tiempo=tiempo+deltat/86400
    
    plt.figure("solucion ME")
    plt.plot(r[1:Nr-1], Pe)
    
    #print("tiempo de simulación:", tiempo)

plt.grid()
```
